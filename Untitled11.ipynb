````python
code = """
import os
import csv
import numpy as np
import trimesh

def process_mesh(file_path, out_dir):
    mesh = trimesh.load(file_path, force='mesh')
    if mesh.is_empty:
        raise ValueError(f"Mesh {file_path} is empty or failed to load.")

    # Extract stats
    n_vertices = int(mesh.vertices.shape[0])
    n_faces = int(mesh.faces.shape[0])
    bbox_min = mesh.bounds[0].tolist()
    bbox_max = mesh.bounds[1].tolist()

    # Normalize mesh: center + uniform scale
    centroid = mesh.centroid
    mesh.vertices -= centroid
    max_extent = float(mesh.extents.max()) if mesh.extents.size else 1.0
    scale = 1.0 / max_extent if max_extent != 0 else 1.0
    mesh.vertices *= scale

    # Save normalized mesh
    base_name = os.path.splitext(os.path.basename(file_path))[0]
    out_path = os.path.join(out_dir, base_name + ".ply")
    mesh.export(out_path)

    return {
        "filename": base_name + ".ply",
        "n_vertices": n_vertices,
        "n_faces": n_faces,
        "bbox_min": bbox_min,
        "bbox_max": bbox_max
    }

def main(input_dir=".", out_dir="./normalized", csv_out="results.csv"):
    os.makedirs(out_dir, exist_ok=True)

    header = ["filename", "n_vertices", "n_faces", "bbox_min", "bbox_max"]
    rows = []

    files = [f for f in os.listdir(input_dir) if f.lower().endswith(".obj")]
    if not files:
        print("No .obj files found in the input directory.")
    for f in files:
        fp = os.path.join(input_dir, f)
        try:
            stats = process_mesh(fp, out_dir)
            rows.append([stats[h] for h in header])
            print(f"Processed {f} -> {stats['filename']}")
        except Exception as e:
            print(f"Failed to process {f}: {e}")

    # Save CSV
    with open(csv_out, "w", newline="") as fh:
        writer = csv.writer(fh)
        writer.writerow(header)
        for r in rows:
            writer.writerow([
                r[0], r[1], r[2],
                "[" + ", ".join(f"{x:.6f}" for x in r[3]) + "]",
                "[" + ", ".join(f"{x:.6f}" for x in r[4]) + "]"
            ])
    print(f"Saved CSV to {csv_out} and normalized meshes to {out_dir}")

if __name__ == "__main__":
    main()
"""

with open("data_prep.py", "w") as f:
    f.write(code)

print("✅ data_prep.py created")
````

````python
code_part2 = r"""
import os
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

def load_data(csv_path='pose_params.csv'):
    if not os.path.exists(csv_path):
        print(f"WARNING: {csv_path} not found. Generating synthetic dataset.")
        n = 200
        X = np.random.randn(n, 6).astype(np.float32)
        coefs = np.array([0.5, -0.2, 0.1, 0.3, 0.0, 0.4], dtype=np.float32).reshape(-1,1)
        y = (X @ coefs).flatten() + 0.1 * np.random.randn(n).astype(np.float32)
        return X, y.reshape(-1,1), [f"feat{i}" for i in range(X.shape[1])]

    df = pd.read_csv(csv_path).dropna()
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

    if 'height' in df.columns:
        feature_cols = [c for c in numeric_cols if c != 'height']
        X = df[feature_cols].values.astype(np.float32)
        y = df['height'].values.astype(np.float32).reshape(-1,1)
    else:
        X = df[numeric_cols].values.astype(np.float32)
        rng = np.random.RandomState(SEED)
        coefs = rng.randn(X.shape[1], 1).astype(np.float32)
        y = (X @ coefs).flatten() + 0.05 * rng.randn(X.shape[0]).astype(np.float32)
        feature_cols = numeric_cols

    return X, y, feature_cols

class MLP(nn.Module):
    def __init__(self, in_dim, hidden_dim=64, out_dim=1):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim//2),
            nn.ReLU(),
            nn.Linear(hidden_dim//2, out_dim)
        )
    def forward(self, x):
        return self.net(x)

def train_model(X, y, epochs=10, lr=1e-3):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    X_t = torch.tensor(X, dtype=torch.float32).to(device)
    y_t = torch.tensor(y, dtype=torch.float32).to(device)

    model = MLP(in_dim=X.shape[1]).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    losses = []
    for epoch in range(epochs):
        optimizer.zero_grad()
        outputs = model(X_t)
        loss = criterion(outputs, y_t)
        loss.backward()
        optimizer.step()
        loss_val = loss.item()
        losses.append(loss_val)
        print(f"Epoch {epoch+1}/{epochs} - Loss: {loss_val:.6f}")

    return model, losses

def plot_loss(losses, out_path='loss_curve.png'):
    plt.figure()
    plt.plot(range(1, len(losses)+1), losses, marker='o')
    plt.xlabel('Epoch')
    plt.ylabel('MSE Loss')
    plt.title('Training Loss Curve')
    plt.grid(True)
    plt.savefig(out_path)
    print(f"Saved {out_path}")

def main():
    X, y, features = load_data()
    print(f"Data: {X.shape[0]} samples, {X.shape[1]} features")
    model, losses = train_model(X, y, epochs=10, lr=1e-3)
    plot_loss(losses)

if __name__ == "__main__":
    main()
"""
with open("ml_baseline.py", "w") as f:
    f.write(code_part2)

print("✅ ml_baseline.py created")
````
